## **Multiple Linear Regression**
*Formula for mlr => y = m1x1 + m2x2 +.... mnxn + c*

```
x1 = [1,2,3,4,5]
x2 = [2,5,3,6,1]

y = [2,5,1,4,7]
```

```
def sum(lst):
  s = 0
  for i in lst:
    s += i
  return s
```

```
 # find mean of y
  y_mean = sum(y) / len(y)
  y_y_mean = [i-y_mean for i in y]

  print(y_mean)
  print(y_y_mean)
```

```
# m1 = E((x1-x1_mean)*(y-y_mean))

def findSlopes(*args,y):
  print(list(args),", y = ",y)
  slopes = []
  for i in args:
    xi_mean = sum(i) / len(i)
    xi_xi_mean = [j-xi_mean for j in i]
    xi_xi_mean_squ = [j*j for j in xi_xi_mean]
    xi_xi_mean_squ_sum = sum(xi_xi_mean_squ)
    xi_y_mean = [xi_xi_mean[j]*y_y_mean[j] for j in range(len(i))]
    xi_y_mean_sum = sum(xi_y_mean)
    slopes.append([xi_y_mean_sum / xi_xi_mean_squ_sum, xi_mean])
  return slopes
```

```
slopes_with_Xmean = findSlopes(x1,x2,y=y)
print(slopes_with_Xmean)
```

```
# find c = y_mean - (m1*x1_mean + m2*x2_mean + ..... mn*xn_mean)

mx = 0

for i in slopes_with_Xmean:
    mx +=  i[0] * i[1]

print(mx)

c = y_mean - mx

print(c)
```

```
# now predict the values

y_pred = []

def predict(length,*args):
  for i in range(length):
    con = c
    for j in range(len(args)):
      con += slopes_with_Xmean[j][0] * args[j][i]
    y_pred.append(con)

  return y_pred
```

```
print(predict(len(y),x1,x2))
```

```
import matplotlib.pyplot as mplot

mplot.scatter(x1,y)
mplot.plot(x1,y_pred)
```

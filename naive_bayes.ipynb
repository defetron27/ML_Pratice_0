```
from sklearn import datasets
 
datasets = datasets.load_iris()
```

```
print(datasets)
```

```
datasets.data
```

```
datas = datasets.data
labels = datasets.target
```

```
print(datas,labels)
```

```
# find the count of labels

# create fun to remove the duplicates

def get_uniques(lst):
  uni = []
  for i in lst:
    if i not in uni:
      uni.append(i)
  return uni
```

```
labels_unique = get_uniques(labels)
print(labels_unique)
```

```
# create fun to count the duplicates

def get_counts(lst,uniques):
  counts = []
  for i in uniques:
    count = 0
    for j in lst:
      if i == j:
        count+=1
    counts.append(count)
  return counts
```

```
labels_counts = get_counts(labels,labels_unique)
print(labels_counts)
```

```
# create fun for get labels total

def get_label_totals(counts,total):
  totals = []
  for i in counts:
    totals.append(i/total)
  return totals
```

```
label_totals = get_label_totals(labels_counts,len(labels))
print(label_totals)
```

```
# inverse the rows and cols in datas in easy access

def get_datas_inverse(lst):
  data = []
  for i in range(len(lst[0])):
    inver = []
    for j in range(len(lst)):
      inver.append(lst[j][i])
    data.append(inver)
  return data
```

```
datas_inv = get_datas_inverse(datas)

print(datas_inv)
print(len(datas_inv))
```

```
# get uniques from the dataset

def get_datas_unique(lst):
  data = []
  for i in range(len(lst)):
    uni = []
    for j in range(len(lst[i])):
      if lst[i][j] not in uni:
        uni.append(lst[i][j])
    data.append(uni)
  return data
```

```
datas_unique = get_datas_unique(datas_inv)

print(datas_unique)

print(len(datas_inv[0]))
print(len(datas_unique[0]))
```

```
for i in range(len(datas_unique)):
  for j in range(len(datas_unique[i])):
    datas_unq = []
    for k in range(len(labels_unique)):
      total = []
      count = 0
      for l in range(len(datas_inv[i])):
        if datas_unique[i][j] == datas_inv[i][l] and labels[l] == labels_unique[k]:
          count+=1
      total.append(count)
    datas_unq.append(total)
  print(datas_unq)
```

```
datas = [["sunny","hot","high","weak"],
["sunny","hot","high","strong"],
["overcast","hot","high","weak"],
["rain","mild","high","weak"],
["rain","cool","normal","weak"],
["rain","cool","normal","strong"],
["overcast","cool","normal","strong"],
["sunny","mild","high","weak"],
["sunny","cool","normal","weak"],
["rain","mild","normal","weak"],
["sunny","mild","normal","strong"],
["overcast","mild","high","strong"],
["overcast","hot","normal","weak"],
["rain","mild","high","strong"],
]

labels = ["no","no","yes","yes","yes","no","yes","no","yes","yes","yes","yes","yes","no"]
```

```
# find the count of labels

# create fun to remove the duplicates

def get_uniques(lst):
  uni = []
  for i in lst:
    if i not in uni:
      uni.append(i)
  return uni
```

```
labels_unique = get_uniques(labels)
print(labels_unique)
```

```
# create fun to count the duplicates

def get_counts(lst,uniques):
  counts = {}
  for i in uniques:
    count = 0
    for j in lst:
      if i == j:
        count+=1
    counts[i] = count
  return counts
```

```
labels_counts = get_counts(labels,labels_unique)
print(labels_counts)
```

```
# create fun for get labels total

def get_label_totals(counts,total):
  totals = {}
  for i,j in zip(labels_counts.keys(),labels_counts.values()):
    totals[i] = j/total
  return totals
```

```
# inverse the rows and cols in datas in easy access

def get_datas_inverse(lst):
  data = []
  for i in range(len(lst[0])):
    inver = []
    for j in range(len(lst)):
      inver.append(lst[j][i])
    data.append(inver)
  return data
```

```
datas_inv = get_datas_inverse(datas)

print(datas_inv)
print(len(datas_inv))
```

```
# get uniques from the dataset

def get_datas_unique(lst):
  data = []
  for i in range(len(lst)):
    uni = []
    for j in range(len(lst[i])):
      if lst[i][j] not in uni:
        uni.append(lst[i][j])
    data.append(uni)
  return data
```

```
datas_unique = get_datas_unique(datas_inv)

print(datas_unique)

print(len(datas_inv[0]))
print(len(datas_unique[0]))
```

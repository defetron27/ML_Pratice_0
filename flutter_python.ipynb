```
import requests
from bs4 import BeautifulSoup as bs

url = "https://flutter.dev/docs/reference/widgets"

page = requests.get(url)
soup = bs(page.content,"html.parser")

results = soup.find_all('header',class_ =  'card-title')
print(results)
```

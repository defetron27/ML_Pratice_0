```
def show_flow_graph(flow_dataframe):
  data1 = flow_dataframe[flow_dataframe.labels == 0]
  data2 = flow_dataframe[flow_dataframe.labels == 1]
  data3 = flow_dataframe[flow_dataframe.labels == 2]

  datas = (data1,data2,data3)
  colors = ('red','green',"yellow")
  groups = ('Iris-setosa','Iris-versicolor','Iris-virginica')

  plt.figure(figsize=(10,7))
  plt.grid(True,linewidth = 0.4,color="black", linestyle="-")

  for data, color, group in zip(datas,colors,groups):
    x = data["petal_width"]
    y = data["petal_length"]
    plt.scatter(x,y,c=color,label=group,)
  plt.legend(loc=2)
  plt.xlabel("Petal Width",color = "white")
  plt.ylabel("Petal Length",color = "white")
  plt.title("Petal Width vs Petal Length",color = "white")
  plt.tick_params(axis='both',colors='white')
```

```
show_flow_graph(flow_dataframe)
```

```
#Train_Test_Split 
def train_test_split(df,test_size):
  # get indices of df
  indices = df.index.tolist()  
  # test if the test_size is float
  if isinstance(test_size,float):
    test_size = round(test_size * len(df))
  # choose random indices from the df
  test_indices = random.sample(population=indices,k=test_size)
  # split the dataset
  test_df = flow_dataframe.loc[test_indices]
  train_df = flow_dataframe.drop(test_indices)
  return train_df,test_df
```

```
random.seed(0) # couldn't change the test data and train data, both are same if again run the code

train_df,test_df = train_test_split(flow_dataframe,test_size = 0.2)

print(train_df,test_df)
```

```
def check_purity(df):
  # get only labels from dataset
  labels_column = df[:,-1];
  # separate uniques
  unique_classes = np.unique(labels_column)

  if len(unique_classes) == 1:
    return True
  else:
    return False

  return False
```
